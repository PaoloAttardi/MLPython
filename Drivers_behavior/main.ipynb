Import the needed libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Look for general insights in the dataset, including the target distribution

```python
df = pd.read_csv('data/sensor_raw.csv', sep=',')
df.head()
df.describe()
targ1 = df.loc[df['Target'] == 1]
targ2 = df.loc[df['Target'] == 2]
targ3 = df.loc[df['Target'] == 3]
targ4 = df.loc[df['Target'] == 4]
print (f"Distribution of the target Class,\n"
       f"     1: {targ1['Target'].count()/df['Target'].count()}\n" 
       f"     2: {targ2['Target'].count()/df['Target'].count()}\n" 
       f"     3: {targ3['Target'].count()/df['Target'].count()}\n" 
       f"     4: {targ4['Target'].count()/df['Target'].count()}")
print(df.info())
print(df.describe().T)
```

```python
# Create a sub plot to showcase all the 4 target data distribution for the single GyroX
fig, axes = plt.subplots(2, 2, figsize=(10, 10), sharey=True, sharex=True)

sns.histplot(data=targ1, x='GyroX', ax=axes[0, 0])
axes[0, 0].set_title(f"Target value: {targ1['Target'].unique()[0]}")

sns.histplot(data=targ2, x='GyroX', ax=axes[0, 1])
axes[0, 1].set_title(f"Target value: {targ2['Target'].unique()[0]}")

sns.histplot(data=targ3, x='GyroX', ax=axes[1, 0])
axes[1, 0].set_title(f"Target value: {targ3['Target'].unique()[0]}")

sns.histplot(data=targ4, x='GyroX', ax=axes[1, 1])
axes[1, 1].set_title(f"Target value: {targ4['Target'].unique()[0]}")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 12))
sns.boxplot(data=df, y='GyroX', ax=axes[0, 0], x='Target')
sns.boxplot(data=df, y='GyroY', ax=axes[0, 1], x='Target')
sns.boxplot(data=df, y='GyroZ', ax=axes[1, 0], x='Target')
sns.countplot(data=df, x='Target', ax=axes[1, 1])
```
